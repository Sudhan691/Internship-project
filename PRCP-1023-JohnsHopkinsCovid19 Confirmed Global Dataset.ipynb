### EDA :

```python
for j in range(1,2):   
    plt.figure(figsize=(40,8))
    for tt in selected_data.columns[(j-1)*7:j*7]:
        confirmed_global = selected_data[[tt]]
        confirmed_global = confirmed_global[confirmed_global[tt]> 0].reset_index(drop=True)
        confirmed_global = confirmed_global.rolling(2).mean().fillna(0)
        plt.plot([str(i) for i in confirmed_global[tt].index],confirmed_global[tt],marker='*',label=tt)
        
    plt.title("COUNTRY LEVEL CONFIRMED CASES GROWTH CURVE")
    plt.legend()
    plt.xlabel('Day')
    plt.ylabel('Confirmed Cases')
    plt.show()
```

### Autocorrelation :

```python
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf

# Plot autocorrelation for each time series
for tt in selected_data.columns:
    confirmed_global = selected_data[[tt]]
    confirmed_global = confirmed_global[confirmed_global[tt] > 0].reset_index(drop=True)
    confirmed_global = confirmed_global.rolling(2).mean().fillna(0)
    
    plt.figure(figsize=(20, 10))
    plot_acf(confirmed_global[tt], lags=40)
    plt.title(f"Autocorrelation for {tt}")
    plt.xlabel('Lag')
    plt.ylabel('Autocorrelation')
    plt.show()
```

### Check for Stationarity and Differencing :

```python
# ADfuller Test to check stationarity
from statsmodels.tsa.stattools import adfuller
import numpy as np

for tt in selected_data.columns:
    confirmed_global = selected_data[[tt]]
    confirmed_global = confirmed_global[confirmed_global[tt] > 0].reset_index(drop=True)
    confirmed_global = confirmed_global.rolling(2).mean().fillna(0)
    
    result = adfuller(confirmed_global[tt])
    print(f'ADF Statistic for {tt}: {result[0]}')
    print(f'p-value for {tt}: {result[1]}')
```

### Determine p and q using ACF and PACF plots :

```python
from statsmodels.graphics.tsaplots import plot_pacf

for tt in selected_data.columns:
    confirmed_global = selected_data[[tt]]
    confirmed_global = confirmed_global[confirmed_global[tt] > 0].reset_index(drop=True)
    confirmed_global = confirmed_global.rolling(2).mean().fillna(0)
    
    plt.figure(figsize=(20, 10))
    plot_acf(confirmed_global[tt], lags=40)
    plt.title(f"ACF for {tt}")
    plt.xlabel('Lag')
    plt.ylabel('Autocorrelation')
    plt.show()

    plt.figure(figsize=(20, 10))
    plot_pacf(confirmed_global[tt], lags=40)
    plt.title(f"PACF for {tt}")
    plt.xlabel('Lag')
    plt.ylabel('Partial Autocorrelation')
    plt.show()
```

### Fit ARIMA Models and Select Best Model :

```python
import statsmodels.api as sm

for tt in selected_data.columns:
    confirmed_global = selected_data[[tt]]
    confirmed_global = confirmed_global[confirmed_global[tt] > 0].reset_index(drop=True)
    confirmed_global = confirmed_global.rolling(2).mean().fillna(0)
    
    best_aic = np.inf
    best_order = None
    best_mdl = None
    
    for p in range(5):
        for d in range(2):
            for q in range(5):
                try:
                    tmp_mdl = sm.tsa.ARIMA(confirmed_global[tt], order=(p, d, q)).fit()
                    tmp_aic = tmp_mdl.aic
                    if tmp_aic < best_aic:
                        best_aic = tmp_aic
                        best_order = (p, d, q)
                        best_mdl = tmp_mdl
                except:
                    continue
    
    print(f"Best ARIMA{best_order} model for {tt} with AIC: {best_aic}")
    print(best_mdl.summary())
```

### Model Diagnostics :

```python
for tt in selected_data.columns:
    confirmed_global = selected_data[[tt]]
    confirmed_global = confirmed_global[confirmed_global[tt] > 0].reset_index(drop=True)
    confirmed_global = confirmed_global.rolling(2).mean().fillna(0)
    
    model = sm.tsa.ARIMA(confirmed_global[tt], order=best_order)
    results = model.fit()
    
    residuals = results.resid
    
    plt.figure(figsize=(10, 5))
    plt.plot(residuals)
    plt.title(f'Residuals of the ARIMA model for {tt}')
    plt.show()
    
    plot_acf(residuals, lags=40)
    plt.title(f'Residuals ACF for {tt}')
    plt.show()
```

### Task -2 Forecasting :

```python
for tt in selected_data.columns:
    confirmed_global = selected_data[[tt]]
    confirmed_global = confirmed_global[confirmed_global[tt] > 0].reset_index(drop=True)
    confirmed_global = confirmed_global.rolling(2).mean().fillna(0)
    
    model = sm.tsa.ARIMA(confirmed_global[tt], order=best_order)
    results = model.fit()
    
    forecast = results.forecast(steps=10)  # Example: forecast next 10 steps
    print(f"Forecast for {tt}:")
    print(forecast)
    
    plt.figure(figsize=(10, 5))
    plt.plot(confirmed_global[tt], label='Observed')
    plt.plot(range(len(confirmed_global), len(confirmed_global) + len(forecast)), forecast, label='Forecast')
    plt.legend()
    plt.title(f'Forecast for {tt}')
    plt.show()
```

### Split the Data :

```python
train_size = int(len(confirmed_global) * 0.8)
train, test = confirmed_global[:train_size], confirmed_global[train_size:]
```

### Fit the Model on Training Set :

```python
model = sm.tsa.ARIMA(train, order=best_order)
fitted_model = model.fit()
```

### Make Predictions on the Test Set :

```python
start = len(train)
end = len(train) + len(test) - 1
predictions = fitted_model.predict(start=start, end=end, typ='levels')
```

###  Calculate Accuracy Metrics :

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(test, predictions)
mse = mean_squared_error(test, predictions)
rmse = np.sqrt(mse)

print(f'Mean Absolute Error: {mae}')
print(f'Mean Squared Error: {mse}')
print(f'Root Mean Squared Error: {rmse}')
```
